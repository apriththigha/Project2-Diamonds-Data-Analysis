*Clean all column names at once*

```python
df.columns = df.columns.str.strip()   
df.columns
```

```python
df.columns = df.columns.str.lower()    
df.columns
```

```python
sns.scatterplot(x='carat weight', y='price', data=df)
plt.title('Price vs Carat Weight')
plt.show()
```

**Histogram (Distribution Plot)**

```python
sns.histplot(df['price'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Diamond Prices')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

**Count Plot (Category Frequency)**

```python
sns.countplot(x='color', hue='color', data=df, palette='viridis', legend=False)
plt.title('Count of Diamonds by Color')
plt.xlabel('Color')
plt.ylabel('Count')
plt.show()
```

**Box Plot (Compare Prices Across Categories)**

```python
sns.boxplot(x='cut', hue ='color', y='price', data=df, palette='pastel')
plt.title('Price vs Cut')
plt.show()
```

**Scatter Plot (Relationship Between Two Numeric Columns)**

```python
sns.scatterplot(x='carat weight', y='price', data=df, color='tomato')
plt.title('Price vs Carat Weight')
plt.show()
```

**Correlation Heatmap**

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

**Pairplot (Mini-EDA Overview)**

```python
sns.pairplot(df[['carat weight', 'depth %', 'table %', 'price']])
plt.show()
```

***Summary***

*histplot	     - See numeric data spread*

*boxplot	     - Compare category vs numeric*

*scatterplot	 - See relationships between two numeric columns*

*heatmap	     - See correlation strength*

*pairplot	     - See overall numeric relationships*
